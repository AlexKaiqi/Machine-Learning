# 3. learning rate（学习率）

学习率会影响代价函数收敛的速度. 一般大的学习率收敛会快些. 但是过大的学习率会导致无法收敛的情况发生.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from util import *
```

```python
# read data
raw_data = pd.read_csv('ex1data2.txt', names=['square', 'bedrooms', 'price'])

# prepare data
data = normalize_feature(raw_data)
X = get_X(data)
y = get_y(data)
epoch = 200
theta = np.zeros(X.shape[1]);
candidate = np.array([1e-4, 1e-3, 1e-2, 1e-1, 1])
print(candidate)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

for alpha in candidate:
    _, cost_data = batch_gradient_descent(theta, X, y, epoch, alpha=alpha)
    ax.plot(np.arange(epoch + 1), cost_data, label=alpha)

ax.set_xlabel('epoch', fontsize=18)
ax.set_ylabel('cost', fontsize=18)
ax.legend(bbox_to_anchor=(1.05, 1))
ax.set_title('learning rate', fontsize=18)
plt.show()
```

```python
_, cost_data = batch_gradient_descent(theta, X, y, epoch, alpha=1.3)
plt.plot(np.arange(epoch + 1), cost_data, label=alpha)
```

```python
_, cost_data = batch_gradient_descent(theta, X, y, epoch, alpha=1.31)
plt.plot(np.arange(epoch + 1), cost_data, label=alpha)
```
